```python
import matplotlib.pyplot as plt
import pandas as pd

file = "speakeasies.csv"
df = pd.read_csv(file)
df.head()
```

```python
df = df[["Rating", "Reviews", "Price(Scale)"]]

# plt.figure(figsize=(200,100))

df.plot.scatter( x = "Reviews", 
                 y = "Rating", 
                 s = 1000,
                 c = "Price(Scale)", 
                 colormap='gnuplot',
                 figsize=(30,10),
                 fontsize = 10,
                 colorbar = True)

plt.savefig('fig.png')
```
